1. String problem. 

Suppose you are given two strings (they may be empty), s1 and s2. You would like to "lace" these strings together, by successively alternating elements of each string (starting with the first character of s1). If one string is longer than the other, then the remaining elements of the longer string should simply be added at the end of the new string. For example, if we lace 'abcd' and 'efghi', we would get the new string: 'aebfcgdhi'.

Write an iterative procedure, called laceStrings(s1, s2) that does this.

Show your test results of  laceStrings(‘abcd’, ‘efghi’)  and
laceStrings(‘efghi’, ‘abcd’)


```python
def laceStrings(s1, s2):
    newString =''
    lenS1 = len(s1)
    lenS2 = len(s2)
    if lenS1 == lenS2:
        for i in range(lenS1):
            newString += s1[i] + s2[i]
    else:
        lenToDo = min(lenS1, lenS2)
        for i in range(lenToDo):
            newString += s1[i] +s2[i]
        if lenS1 > lenS2:
            newString += s1[lenS2:]
        else:
            newString += s2[lenS1:]
    return newString


#testing
print (laceStrings('abcd','efghi'))
print (laceStrings('efghi','abcd'))
#this is  working now.. On to problem 2
```

```python
def uniqueValues(aDict):
    '''
    aDict: a dictionary
    returns: a sorted list of keys that map to unique aDict values, empty list if none
    '''
    result = []
    vals = list(aDict.values())
    for k in aDict:
        if vals.count(aDict[k]) == 1:
            result.append(k)
    return result
#testing If aDict = {1: 1, 3: 2, 6: 0, 7: 0, 8: 4, 10: 0} then your function should return [1, 3, 8]
print(uniqueValues({1: 1, 3: 2, 6: 0, 7: 0, 8: 4, 10: 0}))


#testing If aDict = {1: 1, 2: 1, 3: 1} then your function should return []
print(uniqueValues({1: 1, 2: 1, 3: 1}))


#working as expected! on to probelm 4!....
```
